```python
df["age"]=df["age"].fillna(df["age"].mean())
df["age"]=df["age"].astype("int")
```

```python
df["city"]=df["city"].fillna("Kichha")
df["city"]=df["city"].replace("Kichha", "agra")
# df.info()
df["city"].nunique()
```

```python
df["rating"]=df["rating"].fillna(df["rating"].mean())
```

```python
df["total"]=df["quantity"]*df["price"]*(1-(df["discount"]/100))
```

```python
df
```

```python
df.drop("total", axis=1 , inplace=True)
```

```python
df
```

```python
df.info()
```

```python
a=df.groupby("city")["total"].mean().sort_values(ascending=False).reset_index().round()
a
```

```python
x = a["city"]
y= a["total"]
plt.figure(figsize=(10, 7))

plt.bar(x, y , color="Green")
plt.xlabel("City names" , color="red" )
plt.ylabel("total revenue" ,color="blue")
plt.title("city wise total revenue", color="purple" ,size=20)
plt.grid()
plt.savefig("city wise total revenue.png")
plt.xticks(rotation=45)
plt.yticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
plt.pie(y, labels=x , autopct="%2.0f%%", shadow=True)
plt.axis("equal") 
plt.title("pie chart" , size="30", color="red" , pad=50)
plt.show()
```
