```python
a=df.groupby("city")["age"].mean().sort_values(ascending=False).reset_index().round()
a
```

```python
x=a["city"]
y=a["age"]
plt.figure(figsize=(10, 7))
plt.bar(x, y , color="Green")
plt.xlabel("City names" , color="red" )
plt.ylabel("age mean" ,color="blue")
plt.title("age wise city", color="purple" ,size=20)
plt.grid()
plt.savefig("age wise city user.png")
plt.xticks(rotation=45)
plt.yticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
plt.pie(y, labels=x , autopct="%2.0f%%", shadow=True)
plt.axis("equal") 
plt.title("pie chart" , size="30", color="red" , pad=50)
plt.legend()
plt.show()
```
